# Import libs and data

```python
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')
```

```python
df.info()
```

```python
import seaborn as sns
sns.histplot(df['Price'], bins=30, kde=True, color="steelblue")  # bins=10 oznacza 10 przedziałów
```

# Duplicates?

```python
df.duplicated().sum()
```

# Histograms

```python
for i in df.select_dtypes(include="number").columns:
    if(i!='id'):
        sns.histplot(data=df, x=i, color='steelblue')
        plt.show()
```

```python
for i in df.select_dtypes(include="object").columns:
    if(i!='id'):
        sns.histplot(data=df, x=i, color='steelblue')
        plt.show()
```

# Brand - Price

```python
brands = df[['Brand', 'Price']].groupby(['Brand']).mean()
plt.bar(brands.index, brands['Price'], color='steelblue')
plt.ylim(75,85)
```
